```
 
 #1.List and its default functions
lst=['Diksha','Ajinkya ',10,[1,2,3]]
print("Original list :" ,lst)
 
#append function
lst.append('sneha')
print("List after appending element :" ,lst)
 
#index function
print("index of element 10:" ,lst.index(10))
 
#remove function
lst.remove(10)
print("list after removing element 10:" ,lst)
 
#reverse function
lst.reverse()
print("Reversed list:" ,lst)
 
#clear function
lst.clear()
print("List is cleared :" ,lst)
 
#2.Dictionary and its default functions
 
dict1={'name':'Diksha',
       'Address':'Sangli',
       'mobile no':1234567890,
       'email':'abc@gmail.com'}
 
#get function
dict1.get('name')
 
#pop function
dict1.pop('Address')
print(dict1)
 
#keys function
print(dict1.keys())
 
#copy function
newdict=dict1.copy()
print('original dictionary:',dict1)
print('New dictionary:',newdict)
 
#get function
print('mail of the dict1 is:',dict1.get('email'))
 
 
#3.sets and its default functions
 
fruits={'apple','banana','cherry'}
fruits1={'mango','orange','apple'}
 
#add function
fruits.add('strwbeery')
print(fruits)
 
#copy function
x=fruits.copy()
print(x)
 
#discard function
fruits.discard('apple')
print(fruits)
 
#update function
fruits.update(fruits1)
print(fruits)
 
#clear function
fruits.clear()
print(fruits)
 
#4.Tuple and explore default methods
 
tuple1=(1,3,7,8,7,5,4,6,8,5)
 
#count method
x=tuple1.count(5)
print(x)
 
#index method
x=tuple1.index(8)
print(x)
 
#negative indexing method
print(tuple1[-1])
 
#slicing method
tuple2=('p','a','r','s','o','i','m','g')
 
#print elements 2nd to 4th
print (tuple2[1:4])
 
#elements begining to end
print(tuple2[:])
 
#5.string and explore default methods
 
text="hello,good evening"
 
#capitalize method
x=text.capitalize()
print(x)
 
#center method
x=text.center(40)
print(x)
 
#encode method
x=text.encode()
print(x)
 
#find method
x=text.find('evening')
print(x)
 
#join method
x='*'.join(text)
print(x)
```
